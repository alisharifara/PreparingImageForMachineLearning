```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.data import coins
from sklearn.cluster import AgglomerativeClustering
from sklearn.feature_extraction import image

from scipy.ndimage.filters import gaussian_filter
from skimage.transform import rescale
```

```python
original_coins = coins()
```

```python
original_coins.shape
```

```python
plt.figure(figsize=(12,8))

plt.imshow(original_coins, cmap=plt.cm.gray)

plt.show()
```

```python
smoothened_coins = gaussian_filter(original_coins, sigma=2)
```

```python
plt.figure(figsize=(12,8))

plt.imshow(smoothened_coins, cmap=plt.cm.gray)

plt.show()
```

```python
# scale to 20% of the image
rescaled_coins = rescale(smoothened_coins, 
                         0.2,
                         mode='reflect',
                         multichannel=False,
                         anti_aliasing=False)
rescaled_coins.shape
```

```python
plt.figure(figsize=(12,8))

plt.imshow(rescaled_coins, cmap=plt.cm.gray)

plt.show()
```

```python
#flattening of the image 
X = np.reshape(rescaled_coins, (-1,1))
X.shape
```

```python
# graph of pixel to pixel connections
connectivity = image.grid_to_graph(*rescaled_coins.shape)
print(connectivity)
```

```python
connectivity.shape
```

```python
# to convert into dense array representation
connectivity.toarray()
```

```python
n_clusters = 25
# It works based on bottom-up approach (Hierarchical)
# It recursively merges the pair of clusters, which minimaly increases the *linkage distance
# *Linkage: Criterion to measure distance between clusters
agglo = AgglomerativeClustering(n_clusters=n_clusters,
                               connectivity=connectivity)

agglo.fit(X)
```

```python
# It reshapes the label matrix
# This will allow us to map pixels with corresponding cluster labels
label = np.reshape(agglo.labels_, rescaled_coins.shape)
label
```

```python
label.shape
```

```python
plt.figure(figsize=(12,8))
plt.imshow(rescaled_coins, cmap=plt.cm.gray)

for l in range(n_clusters):
    plt.contour(label==l, colors=[plt.cm.nipy_spectral(l / float(n_clusters)), ])
    
plt.xticks(())
plt.yticks(())
plt.show()
```
